```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

(x_train,y_train),(x_test,y_test)= fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt
n=10
plt.figure(figsize=(20,4))
for i in range(n):
  ax=plt.subplot(2,n,i+1)
  plt.imshow(x_test[i].reshape(28,28))
  plt.gray()
  ax.get_xaxis().set_visible(False)
  ax.get_yaxis().set_visible(False)
plt.show()
plt.close()
```

```python
print("The shape of x_train before scaling",x_train.shape)
```

```python
print("The shape of y_train before scaling",y_train.shape)
```

```python
x_train= x_train.reshape(60000,784)
x_test=x_test.reshape(10000,784)

x_train= x_train.astype('float32')
x_test= x_test.astype('float32')

x_train /=255
x_test /=255

classes=20

y_train=to_categorical(y_train,classes)
y_test=to_categorical(y_test,classes)

print("The shape of x_train after scaling",x_train.shape)
print("The shape of y_train after scaling",y_train.shape)
```

```python
input_size=784
batch_size=200
hidden1=400
hidden2=40
classes=20
epochs=15
```

```python
model=Sequential()

model.add(Dense(hidden1, input_dim=input_size, activation="LeakyReLU"))

model.add(Dense(hidden2, activation="LeakyReLU"))

model.add(Dense(classes, activation="softmax"))

model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer="sgd")

model.summary()
```

```python
from time import time

tic=time()

model.fit(x_train,y_train, batch_size=batch_size, epochs=epochs, verbose=1)

toc=time()

print("Total time for training model= ",toc-tic)
```

```python
from sklearn.metrics import accuracy_score

import numpy as np

import matplotlib.pyplot as plt

y_pred_prob= model.predict(x_test, verbose=0)

y_pred = np.where(y_pred_prob>0.5,1,0)

test_accuracy= accuracy_score(y_pred, y_test)

print("The accuracy is = ",test_accuracy)
```

```python
mask= range(20,50)

x_valid= x_test[0:20]

actual_labels=y_test[0:20]

y_pred_prob_valid= model.predict(x_valid)

y_pred_valid= np.where(y_pred_prob_valid>0.5,1,0)
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]


n=len(x_valid)

plt.figure(figsize=(20,4))

for i in range(n):
  ax=plt.subplot(2,n,i+1)
  plt.imshow(x_valid[i].reshape(28,28))
  plt.gray()
  ax.get_xaxis().set_visible(False)
  ax.get_yaxis().set_visible(False)

  predicted_class= np.argmax(y_pred_prob_valid[i])
  class_name = class_names[predicted_class]

  x=plt.subplot(2,n,i+1+n)
  plt.text(0.5, 0.5, class_name,fontsize=12, ha="center", va="center")
  plt.axis("off")

  plt.show()

  plt.close()
```
